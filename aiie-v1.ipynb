```python
import pandas as pd
from airtable import Airtable

# Configuración de Airtable
# Conectar a Airtable
base_id = 'appNWb7dwg5FVx0Qr'  # ID de la base
api_key = userdata.get('airtable')  # Token de acceso
table_name = "Embedding"  # Nombre de tu tabla

# Conectar a Airtable
airtable = Airtable(base_id, table_name, api_key)
records = airtable.get_all()

# Convertir los datos a un DataFrame
data = pd.DataFrame([record["fields"] for record in records])

# Crear el formato requerido por Pinecone
pinecone_data = []
for _, row in data.iterrows():
    vector = {
        "id": row["id"],  # Asegúrate de que esta columna esté en tu tabla
        "chunk_text": row["chunk_text"],  # Ajusta al nombre de tu columna de texto
        "category": row["category"] if "category" in row else None  # Campo opcional
    }
    pinecone_data.append(vector)

# Guardar los datos en un archivo JSON para revisar antes de subirlos
import json
output_file = "/content/drive/MyDrive/ColabNotebooks/files/pinecone_text_data8.json"
with open(output_file, "w") as f:
    json.dump(pinecone_data, f, indent=2)

print(f"Archivo JSON exportado en: {output_file}")
```

```python
from pinecone import Pinecone
import openai
from google.colab import userdata

# Configurar la API Key de OpenAI
openai.api_key = userdata.get('openai')

# Configurar claves API y crear instancia de Pinecone
pc = Pinecone(
    api_key=userdata.get('pinecone'),
    environment="us-east-1"
)

# Conectar al índice
index_name = "aiie-index"
index = pc.Index(index_name)

# Texto de la consulta
query_text = "How does IE University integrate women and sustainability?"

# Realizar una consulta en Pinecone utilizando el texto directamente
results = index.search_records(
    namespace="values",  # Namespace donde están almacenados los datos
    query={
        "inputs": {"text": query_text},
        "top_k": 16  # Número de resultados relevantes a devolver
    },
    fields=["category", "chunk_text"]  # Campos de metadata que deseas recuperar
)

# Debug: Verificar los resultados obtenidos
print("\nDebugging resultados:")
for hit in results["result"]["hits"]:
    print(f"ID: {hit['_id']}, Score: {hit['_score']}, Category: {hit['fields'].get('category', 'N/A')}")

# Dividir la consulta en keywords
query_keywords = set([word.strip(",.!?").lower() for word in query_text.split()])  # Convertir a set para comparación exacta

# Buscar y puntuar las coincidencias estrictas
filtered_hits = []
for hit in results["result"]["hits"]:
    # Obtener y procesar la categoría del resultado
    categories = set(hit["fields"].get("category", "").lower().split(", "))
    # Verificar si todas las categorías coinciden con las palabras clave
    if categories.issubset(query_keywords) and len(categories.intersection(query_keywords)) > 1:
        filtered_hits.append(hit)

# Extraer el contexto de los mejores resultados
if filtered_hits:
    context = "\n".join(hit["fields"].get("chunk_text", "") for hit in filtered_hits)
else:
    context = "No relevant context found for the query."

# Mostrar resultados
print("Contexto encontrado:")
print(context)

# Mostrar las coincidencias exactas
print("\nResultados filtrados:")
for hit in filtered_hits:
    print(f"ID: {hit['_id']}, Categoría: {hit['fields']['category']}, Texto: {hit['fields']['chunk_text']}")

# Comprobar el prompt generado
print("\nPrompt generado para OpenAI:")
prompt = f"Context: {context}\nQuestion: {query_text}"
print(prompt)

# Usar OpenAI para generar una respuesta
response = openai.ChatCompletion.create(
    model="gpt-4o-2024-08-06",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": prompt}
    ]
)

# Mostrar la respuesta de OpenAI
print("\nRespuesta de OpenAI:")
print(response.choices[0].message["content"])

# Comparar con un prompt sin contexto de Pinecone
print("\nPrueba sin contexto de Pinecone:")
prompt_without_context = f"Question: {query_text}"
response_no_context = openai.ChatCompletion.create(
    model="gpt-4o-2024-08-06",
    messages=[
        {"role": "user", "content": prompt_without_context}
    ]
)
print(response_no_context.choices[0].message["content"])
```

```python
from pinecone import Pinecone
import openai
from google.colab import userdata # Importing the userdata module
import firebase_admin
from firebase_admin import credentials, firestore
import gradio as gr
import json # Importing the json module

# Configurar la API Key de OpenAI
openai.api_key=userdata.get('openai')

# Configurar Pinecone
pc = Pinecone(
    api_key=userdata.get('pinecone'),
    environment="us-east-1"
)
index_name = "aiie-index"
index = pc.Index(index_name)

# Configurar Firebase
# Obtener el secreto desde el llavero de Google Colab
firebase_secret = userdata.get("firebase")  # Obtén el secreto llamado 'firebase'

# Convertir el contenido del secreto (string) a un diccionario JSON
firebase_config = json.loads(firebase_secret)

# Configurar Firebase
cred = credentials.Certificate(firebase_config)  # Usa el secreto en lugar del archivo
firebase_admin.initialize_app(cred)  # Inicializa Firebase
db = firestore.client()  # Cliente Firestore

# Función principal de búsqueda y filtrado
def buscar_respuesta(query_text):
    results = index.search_records(
        namespace="values",
        query={
            "inputs": {"text": query_text},
            "top_k": 16
        },
        fields=["category", "chunk_text"]
    )

    debug_info = []
    for hit in results["result"]["hits"]:
        debug_info.append(f"ID: {hit['_id']}, Score: {hit['_score']}, Category: {hit['fields'].get('category', 'N/A')}")

    query_keywords = set([word.strip(",.!?").lower() for word in query_text.split()])

    filtered_hits = []
    for hit in results["result"]["hits"]:
        categories = set(hit["fields"].get("category", "").lower().split(", "))
        if categories.issubset(query_keywords) and len(categories.intersection(query_keywords)) > 1:
            filtered_hits.append(hit)

    if filtered_hits:
        context = "\n".join(hit["fields"].get("chunk_text", "") for hit in filtered_hits)
    else:
        context = "No relevant context found for the query."

    prompt = f"Context: {context}\nQuestion: {query_text}"
    response = openai.ChatCompletion.create(
        model="ft:gpt-4o-2024-08-06:crowdland:ie-brand:AfrFdtd8",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": prompt}
        ]
    )
    respuesta = response.choices[0].message["content"]

    return respuesta, context, debug_info

def guardar_calificacion(query_text, context, response, rating):
    db.collection("ratings").add({
        "query_text": query_text,
        "context": context,
        "response": response,
        "rating": rating
    })
    return f"Calificación '{rating}' guardada con éxito."

def interfaz(query_text):
    respuesta, context, debug_info = buscar_respuesta(query_text)
    debug_output = "\n".join(debug_info)
    return respuesta, context, debug_output

def calificar(query_text, context, response, rating):
    return guardar_calificacion(query_text, context, response, rating)

with gr.Blocks() as demo:
    query_input = gr.Textbox(label="Ingresa tu consulta:")

    # Botón para generar respuesta
    query_button = gr.Button("Generar Respuesta", elem_id="generate-button")

    # Salidas de respuesta y contexto
    with gr.Column():
        respuesta_output = gr.Textbox(label="Respuesta:", interactive=False)
        context_output = gr.Textbox(label="Contexto:", interactive=False)

    # Botones de calificación en una fila
    with gr.Row():
        good_button = gr.Button("Good")
        regular_button = gr.Button("Regular")
        bad_button = gr.Button("Bad")

    # Salidas de debugging y estado de calificación
    with gr.Column():
        debug_output = gr.Textbox(label="Debugging:", interactive=False)
        calificacion_output = gr.Textbox(label="Estado de Calificación:", interactive=False)

    # Conexión de funciones
    query_button.click(
        interfaz,
        inputs=query_input,
        outputs=[respuesta_output, context_output, debug_output]
    )

    good_button.click(
        calificar,
        inputs=[query_input, context_output, respuesta_output, gr.State("Good")],
        outputs=calificacion_output
    )

    regular_button.click(
        calificar,
        inputs=[query_input, context_output, respuesta_output, gr.State("Regular")],
        outputs=calificacion_output
    )

    bad_button.click(
        calificar,
        inputs=[query_input, context_output, respuesta_output, gr.State("Bad")],
        outputs=calificacion_output
    )

demo.launch()

# CSS Personalizado
gr.Blocks.css = """
#generate-button {
    background-color: #4CAF50; /* Verde */
    color: white;
    border: none;
    padding: 10px 20px;
    font-size: 16px;
    cursor: pointer;
}

#generate-button:hover {
    background-color: #45a049; /* Verde más oscuro al pasar el mouse */
}
"""
```
